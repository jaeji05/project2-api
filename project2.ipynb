## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
!pip3 install requests
import bs4
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed

url = "https://en.wikipedia.org/wiki/Joker:_Folie_%C3%A0_Deux"

# Send a request to the page and get the HTML response
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = bs4.BeautifulSoup(response.text, 'html.parser')

# Remove all tags and get the text content
text_content = soup.get_text()

# Print the text content
print(text_content)
```

```python
#download an image with beautifulsoup and save it in this repository

import os
from urllib.parse import urljoin

# Find all images on the page
images = soup.find_all('img')

# Choose the first image
image_url = images[0]['src']

# Join the relative URL with the base URL
absolute_image_url = urljoin("https://en.wikipedia.org", image_url)

# Download the image
response = requests.get(absolute_image_url)

# Save the image to the current directory
with open('image.jpg', 'wb') as file:
    file.write(response.content)

print("Image downloaded and saved successfully!")
```

```python
#find all the links in a page with beautifulsoup
#print the first 100 characters of ten of these links

# Find all links on the page
links = soup.find_all('a', href=True)

# Print the first 100 characters of ten links
for link in links[:10]:
    print(link['href'][:100])
```

```python
scripts=pd.read_csv('pudding_data.csv')
```

```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags

from bs4 import BeautifulSoup

# Download and clean script text
def download_script(link):
    try:
        response = requests.get(link)
        soup = BeautifulSoup(response.text, 'html.parser')
        text = soup.get_text()[:1000]  # Get first 1000 characters
        return text.strip()  # Remove leading/trailing whitespace
    except Exception as e:
        print(f"Error downloading {link}: {str(e)}")
        return None

# Download script texts
script_texts = scripts['link'].apply(download_script)
```

```python
#add a new column to the df with the text downloaded
#save this new dataframe as "pudding_texts.csv"

# Add new column to the DataFrame
scripts['text'] = script_texts

# Save the updated DataFrame to a new CSV file
scripts.to_csv('pudding_texts.csv', index=False)
```

## Part 4: TMDB database

#### Browse the documentation at https://developer.themoviedb.org/reference/intro/getting-started. Create an account to authenticate

```python
#create a dataset of the movies in theaters now. Include metadata fields you are interested in. 

api_key = "341fc4162a3a1ec3aef15e2fb3659ad2"

# Set API endpoint and parameters
endpoint = f"https://api.themoviedb.org/3/movie/now_playing"
params = {
    "api_key": api_key,
    "language": "en-US",
    "region": "US"
}

# Send GET request
response = requests.get(endpoint, params=params)

# Check if response was successful
if response.status_code == 200:
    # Parse JSON response
    data = response.json()
    
    # Extract relevant metadata
    movies = []
    for result in data["results"]:
        movie = {
            "title": result["title"],
            "overview": result["overview"],
            "release_date": result["release_date"],
            "vote_average": result["vote_average"],
            "poster_path": result["poster_path"],
            "id": result["id"]
        }
        movies.append(movie)
    
    # Create DataFrame
    df = pd.DataFrame(movies)
    
    # Save to CSV
    df.to_csv("movies_in_theaters_now.csv", index=False)
else:
    print("Failed to retrieve data:", response.status_code)
```

```python
#download the movie posters for 10 of these movies and save them to this repository

# Load DataFrame
df = pd.read_csv("movies_in_theaters_now.csv")

# Set poster download directory
poster_dir = "movie_posters"

# Create directory if it doesn't exist
if not os.path.exists(poster_dir):
    os.makedirs(poster_dir)

# Download posters for 10 movies
for index, row in df.head(10).iterrows():
    poster_path = row["poster_path"]
    poster_url = f"https://image.tmdb.org/t/p/w500{poster_path}"
    response = requests.get(poster_url)
    
    if response.status_code == 200:
        # Save poster to file
        filename = f"{row['title']}.jpg"
        filepath = os.path.join(poster_dir, filename)
        with open(filepath, "wb") as f:
            f.write(response.content)
        print(f"Downloaded poster for {row['title']}")
    else:
        print(f"Failed to download poster for {row['title']}: {response.status_code}")
```
